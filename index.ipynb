```python
# Find Second Largest

import numpy as np

arr= np.array([1,2,3,4,5,6,7,9,10])
max1=arr[0]
max2=arr[1]

for i in range(0, len(arr)):
  if max1<arr[i]:
    max1=arr[i]
  
for i in range(1, len(arr)-1):
  if max2<arr[i]:
    max2=arr[i]

if max2<max1:
  print("second lagest element", max2)
else:
  print(max1)
```
