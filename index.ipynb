```python
# Reverse Array
import numpy as np
arr= np.array([1,2,3,4,5,6,7,9,10])

for i in arr[::-1]:
  print(i)
```

```python
# Find Second Largest



arr= np.array([1,2,3,4,5,6,7,9,10])
max1=arr[0]
max2=arr[1]

for i in range(0, len(arr)):
  if max1<arr[i]:
    max1=arr[i]
  
for i in range(1, len(arr)-1):
  if max2<arr[i]:
    max2=arr[i]

if max2<max1:
  print("second lagest element", max2)
else:
  print(max1)
```

```python
# Two Sum

arr= np.array([1,2,3,4,5,6,7,9,10])

target=6
for i in range(0, len(arr)):
  for j in range(i+1, len(arr)):
    if arr[i]+arr[j]==target:
      print(i,j)
```

```python
# Move Zeroes to End – Python Code (Optimal)

arr= np.array([1,0,2,0,3])
count=0

for i in range(len(arr)):
  if arr[i] !=0:
    arr[count]= arr[i]
    count +=1
while count < len(arr):
  arr[count]=0
  count +=1
arr
```
